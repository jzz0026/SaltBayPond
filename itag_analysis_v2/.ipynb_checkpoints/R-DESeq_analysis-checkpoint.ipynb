```r
options(width=110)

library('biom')
library('vegan')
library('DESeq2')
```

## Run DESeq2

```r
# load biom file
OTU.biom <- read_biom("../OTU_table.biom") #'OTU_table.biom'

# Extract data matrix (OTU counts) from biom table
OTU <- as.matrix(biom_data(OTU.biom))
    
# transpose so that rows are samples and columns are genera
OTU <- t(OTU)

# load mapping file
map <- read.table('../Salt_pond_ALL_Meta_whh.txt', sep='\t', comment='', head=T, row.names=3)

# find the overlapping samples
common.ids <- intersect(rownames(map), rownames(OTU))

# get just the overlapping samples
OTU <- OTU[common.ids,]
map <- map[common.ids,]
Restoration <- map$Restoration

dds <- DESeqDataSetFromMatrix(t(OTU), DataFrame(Restoration), design= ~ Restoration ) ## Count table should be tranposed
Faimliy_DESeq <- DESeq(dds, test="Wald", fitType="parametric")

#res <- results(Faimliy_DESeq,contrast=c("Restoration", Treat1, Treat2 )) # "Salt_Pond", "Historic"
res_SH <- lfcShrink(Faimliy_DESeq, contrast=c("Restoration", "Salt_Pond", "Historic"))
res_SR <- lfcShrink(Faimliy_DESeq, contrast=c("Restoration", "Salt_Pond", "Restored"))



#Faimliy_res <- summary(Faimliy_DESeq)
OTU_res_SH <- summary(res_SH)
OTU_res_SR <- summary(res_SR)
#OTU_DESeq
res_SH
```

```r
summary_res_SH <- as.matrix(res_SH)
write.csv(summary_res_SH, file = "OTU_difference_res_SH.csv")
summary_res_SR <- as.matrix(res_SR)
write.csv(summary_res_SR, file = "OTU_difference_res_SR.csv")
```

```r
# Plot results
options(repr.plot.width=4, repr.plot.height=4)
plotMA(res_SH) #, 
plotMA(res_SR)
```

## Get DESeq2 variance stablized data set

```r
## get log2 counts
OTU_dds_vs <- estimateSizeFactors(dds)
OTU_dds_vs <- estimateDispersions(OTU_dds_vs)
OTU_dds_VST <- getVarianceStabilizedData(OTU_dds_vs)  # OTU_phy2des_VST
head(OTU_dds_VST)
```

```r
## get total counts
OTU_vs_counts <- counts(OTU_dds_vs, normalized=TRUE)     # get counts                       #head(KO_vs_counts) 
head(OTU_vs_counts)
write.csv(OTU_vs_counts, file = "OTU_counts_nomalized.csv")

## nomalized to 10^6
tpm <- (round(OTU_vs_counts/colSums(OTU_vs_counts)[col(OTU_vs_counts)] *1e6,2))
head(tpm)
write.csv(tpm, file = "OTU_counts_nomalized_relative.csv")
```

```r
DSq_cntrstF = function(pwd, Treat1, Treat2) { 
    # load biom file
    family.biom <- read_biom(pwd) #'Family_table.biom'

    # Extract data matrix (family counts) from biom table
    family <- as.matrix(biom_data(family.biom))
    
    # transpose so that rows are samples and columns are genera
    family <- t(family)

    # load mapping file
    map <- read.table('Salt_pond_ALL_Meta_whh.txt', sep='\t', comment='', head=T, row.names=3)

    # find the overlapping samples
    common.ids <- intersect(rownames(map), rownames(family))

    # get just the overlapping samples
    family <- family[common.ids,]
    map <- map[common.ids,]
    Restoration <- map$Restoration

    dds <- DESeqDataSetFromMatrix(t(family), DataFrame(Restoration), design= ~ Restoration ) ## Count table should be tranposed
    Faimliy_DESeq <- DESeq(dds, test="Wald", fitType="parametric")

    #res <- results(Faimliy_DESeq,contrast=c("Restoration", Treat1, Treat2 )) # "Salt_Pond", "Historic"
    res <- lfcShrink(Faimliy_DESeq, contrast=c("Restoration", Treat1, Treat2))

    return(res)
}
```
